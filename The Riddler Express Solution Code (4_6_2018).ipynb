```python
import pandas as pd
import numpy as np
dates = pd.read_csv('date.csv')
dates['first'] = dates.Date.str.extract('([^/]+)/[^/]+/[^/]+').astype(float)
dates['second'] = dates.Date.str.extract('[^/]+/([^/]+)/[^/]+').astype(float)
dates['third'] = dates.Date.str.extract('[^/]+/[^/]+/([^/]+)').astype(float)

answers =[]
count = 0

for i in dates.Date:
    if int(dates['third'][count]) == int(dates['second'][count])*int(dates['first'][count]):
        answers.append(i)
    count += 1
```
